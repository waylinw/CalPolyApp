```python
import urllib2
from bs4 import BeautifulSoup
```

```python
#FOR all colleges
to_save = {"Department": [], "Course": [], "Section": [], "Instructor": [], "Time": []}
sources = ["http://schedules.calpoly.edu/depts_10-CAGR_curr.htm", "http://schedules.calpoly.edu/depts_20-CAED_curr.htm", 
          "http://schedules.calpoly.edu/depts_40-OCOB_curr.htm", "http://schedules.calpoly.edu/depts_48-CLA_curr.htm", 
          "http://schedules.calpoly.edu/depts_52-CENG_curr.htm", "http://schedules.calpoly.edu/depts_76-CSM_curr.htm", 
          "http://schedules.calpoly.edu/depts_99-ALL_curr.htm"]

for source in sources:
    opener = urllib2.build_opener()
    our_url = opener.open(source).read()
    soup = BeautifulSoup(our_url)
    b = BeautifulSoup(str(soup.find_all("table")[0]))
    
    
    latest_prof = ""
    for tr in b.find_all('tr'):
        tds = BeautifulSoup(str(tr)).find_all('td')
        
        #to get rid of empty no finds with tr's
        if tds:
            #some have no prof and some have no coursename (prof without a course)
            prof_name = BeautifulSoup(str(tds)).findAll("td", {"class": "personName"})
            
            if prof_name:
                #remember latest prof is necessary because only a prof's first class has name on it; if it is not there
                #assume last mentioned prof is the one for the class so we are saving it here whenever a new prof name 
                #is seen for later reference
                latest_prof = str(BeautifulSoup(str(prof_name)).get_text())[1:-1]
            course_name = BeautifulSoup(str(tds)).findAll("td", {"class": "courseName"})
            
            if course_name:
                course_name = str(BeautifulSoup(str(course_name)).get_text())[1:-1]
            
            #getting these fields as well
            section_num = str(BeautifulSoup(str(BeautifulSoup(str(tds)).findAll("td", {"class": "courseSection"}))).get_text())
            day = str(BeautifulSoup(str(BeautifulSoup(str(tds)).findAll("td", {"class": "courseDays"}))).get_text())
            start_time = str(BeautifulSoup(str(BeautifulSoup(str(tds)).findAll("td", {"class": "startTime"}))).get_text())
            end_time = str(BeautifulSoup(str(BeautifulSoup(str(tds)).findAll("td", {"class": "endTime"}))).get_text())
            
            # unicode mess cleanup
            section_num = section_num.replace("\\xa0", "").replace("\\", "")
            day = day.replace("\\xa0", "N/A").replace("\\", "")
            start_time = start_time.replace("\\xa0", "N/A").replace("\\", "")
            end_time = end_time.replace("\\xa0", "N/A").replace("\\", "")
            
            # getting rid of brackets
            if section_num:
                section_num = section_num[1:-1]
            if day:
                day = day[1:-1]
            if start_time:
                start_time = start_time[1:-1]
            if end_time:
                end_time = end_time[1:-1]

            # only proceed if there is a course name because no time, day, dept etc if there is not
            if course_name:
                #combine for one time field
                final_time = day + " " + start_time + " - " + end_time
                
                #separate dept and such as well
                dept_num = course_name.split(" ")
                
                to_save["Department"].append(dept_num[0])
                to_save["Course"].append(dept_num[1])
                to_save["Section"].append(section_num)
                to_save["Instructor"].append(latest_prof)
                to_save["Time"].append(final_time)
```

```python
import pandas as pd
save_csv = pd.DataFrame(to_save)
save_csv = save_csv.sort(['Department', 'Course', 'Section'])
cols = save_csv.columns.tolist()
cols = ['Department', 'Course', 'Section', 'Instructor', 'Time']
save_csv = save_csv[cols]
save_csv.to_csv('/Users/gurjeevan/Documents/school/jr/sr_proj/parsing/clean_classes.csv')
```
